Debido a la situacion inflacionaria en Argentina se opta por agregar informacion sobre el dolar blue para no tener el sesgo marcado por la inflacion. Para esto obtengo datoos historico de la pagina https://es.investing.com/currencies/usd-arsb-historical-data 
Y uso estos nuevos datos

```python
dolar = pd.read_csv ('Datos históricos USD_ARSB (1).csv')
```

```python
dolar
```

Retiro columnas que no me sirven para el analisis y ajusto el nombre

```python
dolar.drop(columns=(['Apertura', 'Máximo' , 'Mínimo', 'Vol.','% var.']),inplace=True)
```

Cambio el nombre de la columna Ultimo a Valor dolar y lo convierto en float

```python
dolar.rename(columns={'Último':'Valor_dolar'}, inplace=True)
dolar['Valor_dolar'] = dolar['Valor_dolar'].str.replace('.', '').str.replace(',', '.')
dolar['Valor_dolar'] = dolar['Valor_dolar'].astype(float)
```

```python
dolar.info()
```

Convierto fecha a datetime

```python
dolar['Fecha'] = pd.to_datetime(dolar['Fecha'], format='%d.%m.%Y')
```

Creo Año y Trimestre

```python
dolar['Año'] = dolar['Fecha'].dt.year
dolar['Trimestre'] = dolar['Fecha'].dt.to_period('Q').dt.quarter
```

```python
dolar
```

```python
df_dolar = dolar.groupby(['Año', 'Trimestre']).agg({'Valor_dolar': 'mean'})
```

```python
df_dolar
```

Agrupo a Ingresos

```python
ingresos_dolar = pd.merge(df_ingresos, df_dolar, on=['Año', 'Trimestre'], how='left')
```

```python
ingresos_dolar 
```

Calculo el valor en dolares

```python
ingresos_dolar['Ingresos en USD'] = ingresos_dolar ['Ingresos (miles de pesos)'] / ingresos_dolar ['Valor_dolar']
```

```python
ingresos_dolar 
```

Guardo en un nuevo csv

```python
ingresos_dolar.to_csv('data/ingresos_dolar.csv', index=False)
```
